```python
import numpy as np
from random import uniform

def tanh(x):
    return np.tanh(x)

def tanh_deriv(x):
    return 1.0 - np.tanh(x)**2
```

```python
def pred(w_x, x, w_b):
    return (100 * tanh(w_x * x + w_b))
```

```python
def pred_deriv_w_x(w_x, x, w_b):
    return (100 * tanh_deriv(w_x * x + w_b) * x)
```

```python
def pred_deriv_w_b(w_x, x, w_b):
    return (100 * tanh_deriv(w_x * x + w_b))
```

```python
def func(x):
    return (2 * x + 8)
```

```python
def loss(w_x, x, w_b):
    l = 0
    for item in x:
        sq = (pred(w_x, x, w_b) - func(x)) * (pred(w_x, x, w_b) - func(x))
        l = l + sq
    return l
```

```python
def loss_deriv_w_x(w_x, x, w_b):
    l = 0
    for item in x:
        sq = 2 * (pred(w_x, x, w_b) - func(x)) * pred_deriv_w_x(w_x, x, w_b)
        l = l + sq
    return l
```

```python
def loss_deriv_w_b(w_x, x, w_b):
    l = 0
    for item in x:
        sq = 2 * (pred(w_x, x, w_b) - func(x)) * pred_deriv_w_b(w_x, x, w_b)
        l = l + sq
    return l
```

```python
epochs = 1000
step_w_x = 0.01
step_w_b = 0.01
x = np.arange(40)

w_x = uniform(0, 10)
w_b = uniform(0, 10)
```

```python
for i in range(epochs):
    l = loss(w_x, x, w_b)
    print "Loss = " + str(l)
    w_x = w_x - step_w_x * loss_deriv_w_x(w_x, x, w_b)
    print "w_x = " + str(w_x)
    w_b = w_b - step_w_b * loss_deriv_w_b(w_x, x, w_b)
    print "w_b = " + str(w_b)    
```
