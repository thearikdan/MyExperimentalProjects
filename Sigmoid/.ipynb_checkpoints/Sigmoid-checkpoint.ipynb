```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = np.arange(-10., 10., 0.2)
```

```python
def logistic(x):
    a = []
    for item in x:
        a.append(1 / (1 + math.exp(-item)))
    return a
```

```python
log = logistic(x)
```

```python
plt.plot(x,log)
plt.show()
```

```python
def logistic_derivative(x):
    a = []
    for item in x:
        s = 1 / (1 + math.exp(-item))
        a.append(s * (1 - s))
    return a
```

```python
log_der = logistic_derivative(x)
```

```python
plt.plot(x,log_der)
plt.show()
```

```python
tanh = np.tanh(x)
```

```python
plt.plot(x,tanh)
plt.show()
```

```python
def tanh_deriv(x):
    return 1.0 - np.tanh(x)**2
```

```python
tanh_der = tanh_deriv(x)
```

```python
plt.plot(x,tanh_der)
plt.show()
```

```python
def relu(x):
    return np.maximum(x, 0)
```

```python
rel = relu(x)
```

```python
plt.plot(x,rel)
plt.show()
```

```python
def relu_deriv(x):
    return (x > 0)
```

```python
rel_der = relu_deriv(x)
```

```python
plt.plot(x,rel_der)
plt.show()
```
