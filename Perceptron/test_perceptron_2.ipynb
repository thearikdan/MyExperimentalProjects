```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from random import uniform
import math
from perceptron_2 import Perceptron_2
```

```python
w = uniform(-10, 10)
b = uniform(-10, 10)

p = Perceptron_2(w, b)
x =np.arange(-10,10,0.1)

y = p.out(x)
```

```python
fig,(out) = plt.subplots(1,1, sharex = False, sharey = False)

out.set_xlabel('x')
out.set_ylabel('out')
out.set_xlim(-10, 10)
out.set_ylim(-1, 1)
out.scatter(x, y, color = 'b')
```

```python
w_deriv = p.w_deriv(x)

fig,(w_der) = plt.subplots(1,1, sharex = False, sharey = False)

w_der.set_xlabel('x')
w_der.set_ylabel('w_der')
w_der.set_xlim(-10, 10)
w_der.set_ylim(-5, 5)
w_der.scatter(x, w_deriv, color = 'b')
```

```python
b_deriv = p.b_deriv(x)

fig,(b_der) = plt.subplots(1,1, sharex = False, sharey = False)

b_der.set_xlabel('x')
b_der.set_ylabel('b_der')
b_der.set_xlim(-10, 10)
b_der.set_ylim(-2, 2)
b_der.scatter(x, b_deriv, color = 'b')
```
