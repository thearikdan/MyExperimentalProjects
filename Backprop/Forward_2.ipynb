```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from random import uniform
import math
```

```python
def inp(w, x, b):
    return w * x + b


def out(t):
    return np.tanh(t)


def neuron(w, x, b):
    return out(inp(w, x, b))


def pred(x, w11_x, w11_b, w12_x, w12_b, w21_x, w22_x, w2_b):
    return (w21_x * neuron(w11_x, x, w11_b) + w22_x * neuron(w12_x, x, w12_b) + w2_b)


def func(x):
    return np.sin(math.pi * x)

```

```python
x = np.arange(-10, 10, 0.2)
print x

#y = func(x)
#print y

w11_x = -40
w11_b = 100
w12_x = -60
w12_b = 100

w21_x = 0.8
w22_x = 0
w2_b  = 0

t1 = pred(x, w11_x, w11_b, w12_x, w12_b, w21_x, w22_x, w2_b)
print t1

w21_x = 0
w22_x = -0.8

t2 = pred(x, w11_x, w11_b, w12_x, w12_b, w21_x, w22_x, w2_b)
print t2

w21_x = 0.8
w22_x = -0.8

t3 = pred(x, w11_x, w11_b, w12_x, w12_b, w21_x, w22_x, w2_b)
print t3

```

```python
fig,(ax1) = plt.subplots(1,1, sharex = False, sharey = False)

ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_xlim(-10, 10)
ax1.set_ylim(-2, 2)
ax1.set_xlabel("X")
#ax1.plot(x, y, color = 'r')

ax1.plot(x, t1, color = 'g')
ax1.plot(x, t2, color = 'r')
ax1.plot(x, t3, color = 'b')

```
